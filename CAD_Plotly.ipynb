```python
import plotly.express as px

fig = px.pie(df, names='Cath', color_discrete_sequence=px.colors.sequential.RdBu,title="Distribution of gender")
fig.show()
```

```python
labels = df['VHD'].unique()
values = df['VHD'].value_counts()
irises_colors = ['rgb(33, 75, 99)', 'rgb(79, 129, 102)', 'rgb(151, 179, 100)',
                 'rgb(175, 49, 35)', 'rgb(36, 73, 147)']
# pull is given as a fraction of the pie radius
fig = go.Figure(data=[go.Pie(labels=labels, values=values, pull=[0, 0, 0, 0.2],marker_colors=irises_colors)])
fig.update(layout_title_text='Distribution of VHD')
fig.show()
```

```python
labels = df['Current Smoker'].unique()
values = df['Current Smoker'].value_counts()
cafe_colors =  ['rgb(146, 123, 21)', 'rgb(177, 180, 34)', 'rgb(206, 206, 40)',
                'rgb(175, 51, 21)', 'rgb(35, 36, 21)']
# pull is given as a fraction of the pie radius
fig = go.Figure(data=[go.Pie(labels=labels, values=values, pull=[0, 0.1],marker_colors=cafe_colors)])
fig.update(layout_title_text='Distribution of Current Smoker numbers')
fig.show()
```

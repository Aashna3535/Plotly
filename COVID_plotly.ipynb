```python
df = pd.read_csv("covid_19_data (1).csv")
df
```

```python
data = df.copy()
```

```python
data.isnull().sum()
```

```python
data = data.dropna(how = "any")
```

```python
#Number of records having missing value
data.isnull().values.any()
```

```python
#Checking duplicate rows
dupval = data.duplicated()
print(dupval.sum())
```

```python
#import matplotlib.pyplot as plt
#import seaborn as sns
#sns.set()
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
```

```python
fig = px.histogram(data, x='Country/Region',y='Confirmed',color_discrete_sequence=['green'])
fig.show()
```

```python
fig = px.histogram(data, x='Country/Region',y='Deaths',color_discrete_sequence=['teal'])
fig.show()
```

```python
fig = px.histogram(data, x='Country/Region',y='Recovered',color_discrete_sequence=['lightpink'])
fig.show()
```

```python
labels = data['Country/Region'].unique()
size= data['Country/Region'].value_counts()
explode = [0, 0.2]
irises_colors = ['rgb(33, 75, 99)', 'rgb(79, 129, 102)', 'rgb(151, 179, 100)',
                 'rgb(175, 49, 35)', 'rgb(36, 73, 147)']
fig = go.Figure(data=[go.Pie(labels=labels, values=size, marker_colors=irises_colors)])
fig.show()
```

```python
fig = px.scatter(data, x='Confirmed',y='Recovered',color="Deaths")
fig.show()
```

```python
some_countries = ('US','India')
data_countries = data.loc[data['Country/Region'].isin(some_countries) & data['Confirmed']]

fig = px.violin(data_countries, y="Confirmed", x='Country/Region',color_discrete_sequence=['purple'])
fig.show()
```

```python

some_countries = ('Brazil' , 'US', 'India', 'Italy', 'Canada')
data_countries = df.loc[df['Country/Region'].isin(some_countries) & df['Recovered']]

fig = px.bar(data_countries,x="Country/Region",y="Recovered",color_discrete_sequence=['red'])
fig.show()
```

```python
some_states = ('Amazonas', 'Limburg', 'Punjab')
data_states = df.loc[df['Province/State'].isin(some_states) & df['Confirmed']]
fig=px.box(data_states,x="Province/State",y="Confirmed",color_discrete_sequence=['orange'])
fig.show()
```

```python
some_states = ('Amazonas', 'Limburg', 'Punjab')
data_states = df.loc[df['Province/State'].isin(some_states) & df['Confirmed']]
fig= px.line(data_states,x="Province/State",y="Confirmed",color_discrete_sequence=['orange'], title='Line plot of overall Confirmed cases')
fig.show()
```

```python
import plotly.figure_factory as ff
colorscale = ['#7A4579', '#D56073', 'rgb(236,158,105)', (1, 1, 0.2), (0.98,0.98,0.98)]

fig = ff.create_2d_density(
    data['Confirmed'], data['Recovered'], colorscale=colorscale,
    hist_color='rgb(255, 237, 222)', point_size=3,
)
fig.show()
```

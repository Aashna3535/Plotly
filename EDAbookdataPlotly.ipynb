```python
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
```

```python
trace = go.Heatmap(
   x = df['User Rating'],
   y = df['Reviews'],
   z = df['Year'],
   type = 'heatmap',
   colorscale = 'Viridis'
)
data = [trace]
fig = go.Figure(data= data )
fig.show()
```

```python
labels = ['Non Fiction','Fiction'] 
size= df['Genre'].value_counts()
colors = plt.cm.cool(np.linspace(0, 1, 5))
explode = [0, 0.2]
fig = go.Figure(data=[go.Pie(labels=labels, values=size, pull=[0, 0.1])])
fig.show()
```

```python
labels = df['User Rating'].unique()
size= df['User Rating'].value_counts()
colors = plt.cm.cool(np.linspace(0, 1, 5))
explode = [0, 0.2]
fig = go.Figure(data=[go.Pie(labels=labels, values=size, pull=[0, 0.1])])
fig.show()
```

```python
# plotting the histogram
fig = px.histogram(df, x="User Rating") 
  
# showing the plot
fig.show()
```

```python
# plotting the histogram
fig = px.histogram(df, x="Reviews",color_discrete_sequence=['mediumorchid']) 
  
# showing the plot
fig.show()
```

```python
# plotting the histogram
fig = px.histogram(df, x="Price",color_discrete_sequence=['indianred']) 
  
# showing the plot
fig.show()
```

```python
# plotting the histogram
fig = px.histogram(df, x="Author",color_discrete_sequence=['teal']) 
  
# showing the plot
fig.show()
```

```python
# plotting the histogram
fig = px.scatter(df, y="Reviews",x='Price', color="Genre",size="User Rating") 
  
# showing the plot
fig.show()
```

```python
# plotting the bar chart
fig = px.bar(df, x="Year", y="Reviews",color="Genre") 
  
# showing the plot
fig.show()
```

```python
# plotting the box chart
fig = px.box(df, x="Genre", y="Year") 
  
# showing the plot
fig.show()
```

```python
# plotting the violin chart
fig = px.violin(df, x="Genre", y="Reviews")
  
# showing the plot
fig.show()
```

```python
# plotting the figure
fig = px.line_3d(df, x="Reviews", y="Price", 
                 z="User Rating", color="Genre") 
  
fig.show()
```

```python
# plotting the histogram
fig = px.scatter_3d(df, y="Reviews",x='Price' ,z="Year" , color="Genre",size="User Rating") 
  
# showing the plot
fig.show()
```

```python
#Plotting parallel categories
fig = px.parallel_categories(df, color="User Rating", color_continuous_scale=px.colors.sequential.Inferno)
fig.show()
```

```python
#Plotting parallel categories
fig = px.parallel_categories(df, color="Year", color_continuous_scale=px.colors.sequential.Inferno)
fig.show()
```

```python
#Plotting Area chart
fig = px.area(df, x="Year", y="Reviews", color="Genre", line_group="User Rating")
fig.show()
```

```python
#Pllotting 2D Histogram contours
fig = px.density_contour(df, x="User Rating", y="Price",color="Genre",marginal_x="rug", marginal_y="histogram")
fig.show()
```
